```python
# Import Libraries

from transformers import DistilBertForQuestionAnswering, DistilBertTokenizer
import torch
```

```python
# Load the DistilBERT Model and Tokenizer
model_name = "distilbert-base-uncased-distilled-squad"  # A model fine-tuned for question answering
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DistilBertForQuestionAnswering.from_pretrained(model_name)
```

```python
# Define the Truncated Context
context = """
- Cybersecurity: Cybersecurity is the practice of protecting systems, networks, and data from cyber threats, including unauthorized access, damage, and attacks.

- Definition of Cyber Attack: A cyber attack is an attempt to gain unauthorized access to a computer system, network, or digital device with the intent to cause harm or steal information.

- Types of Cyber Attacks and Prevention Measures:
  - Phishing: Deceiving individuals into providing sensitive information.
    - Prevention: Educate users; use email filters.

  - Ransomware: Encrypts files, demanding ransom for decryption.
    - Prevention: Regular data backups; keep software updated.

  - DDoS: Overwhelms a target with excessive traffic.
    - Prevention: Rate limiting and traffic filtering; DDoS protection services.

  - Malware: Includes viruses and Trojans designed to damage systems.
    - Prevention: Use antivirus software; avoid untrusted downloads.

  - Insider Threats: Threats from individuals within the organization.
    - Prevention: Monitor user activities; implement strict access controls.
"""
```

```python
# Create a Function to Answer Questions
def answer_question(question, context):
    inputs = tokenizer.encode_plus(question, context, return_tensors="pt")

    # Get the model's predictions
    with torch.no_grad():
        outputs = model(**inputs)

    # Find the start and end scores
    start_scores = outputs.start_logits
    end_scores = outputs.end_logits

    # Get the most likely start and end of the answer
    start_index = torch.argmax(start_scores)
    end_index = torch.argmax(end_scores) + 1

    # Convert tokens to string
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][start_index:end_index]))
    return answer
```

```python
# Create the Chatbot Interface
def chatbot():
    print("Welcome to the Cybersecurity Chatbot! (Type 'exit' to stop)")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Chatbot: Goodbye!")
            break
        answer = answer_question(user_input, context)
        print(f"Chatbot: {answer}")
```

```python
# Start the chatbot
chatbot()
```
